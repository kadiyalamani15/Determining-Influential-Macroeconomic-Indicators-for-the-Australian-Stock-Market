```python
# Function to perform iterations of PCA to find out eigen values & no. of components

def pca_components(df):
  
  # Scaling data as the values should be closer to perform PCA
  scaler = StandardScaler()
  scaled_data = scaler.fit_transform(df)
  pca = PCA()
  dataset_pca = pca.fit(scaled_data).transform(scaled_data)
  eigen_values = pca.explained_variance_ # Eigen Values

  components = 0
  # K1-Kaiser Criterion: Eigen Values > 1 gives the no. of components
  for i in eigen_values:
    if i >= 1:
      components += 1
  
  return (eigen_values, components, scaled_data)
```

```python
# Function to perform iterations of PCA VARIMAX(Orthogonal Rotation) to find out the component loadings and cummulative variance
# No. of factors recieved from pca_components function is passed here
def varimax(factors, scaled_data):

  labels = ['Factors' + str(i) for i in range(1, factors+1)] # Column Names

  fa = FactorAnalyzer()
  fa.set_params(n_factors=factors, method='principal', rotation='varimax')
  fa.fit(scaled_data)
  # Loadings DataFrame
  loadings_data = pd.DataFrame(fa.loadings_, index=df.columns, columns=labels)
  loadings_data = loadings_data.round(3) # Rounding the values
  # Deleting values whose absolute value is less than 0.5, which helps in grouping variables to corresponding components
  loadings_data[loadings_data.abs() < 0.5] = None 
  loaded_variables = loadings_data.isnull().sum(axis=1) # Finding total null values per variable
  
  cum_variance = fa.get_factor_variance() # Cummulative Variance

  for loc, val in enumerate(loaded_variables):
    # Each factor should be loaded onto only one component, failed to do so, they must be deleted
    if val != (factors -1):
      statement = f"Reiterate PCA test and drop variable {loaded_variables.index[loc]}"
      break
    else:
      statement = "Everything looks good, proceed with ADF"
      continue


  return (loadings_data, cum_variance, statement)
```

```python
# PCA and VARIMAX iteration 1

# PCA
e_v, factors, scaled_data = pca_components(df)
print(f'No. of factors: {factors}')
# VARIMAX
loadings_data, cum_variance, res = varimax(factors, scaled_data)
print(res)
```

```python
# Preparing DataFrame of the Reduced Dimension for ADF test

df_copy = pd.read_excel('79_indices_agg.xlsx')
data_for_adf = df_copy[['DATE','AUSCCON', 'AUSLFPR', 'AUSG', 'AUSCA', 'AUSMP', 'AUSGAGR',
       'AUSPTEMP', 'AUSPPIC', 'AUSGBVL']]
data_for_adf.set_index('DATE', inplace=True)
data_for_adf
```

```python
from statsmodels.tsa.stattools import adfuller
%matplotlib inline

# Function to present graphs and results of ADF
def adf_res(series, loc):


  result = adfuller(series, autolag='AIC')
  print("{:-^150}".format(''))
  print('\n')

  # Plotting the series
  series.plot(figsize=(10,5), legend=None, title=series.name)
  plt.show()

  print('\n')
  statement = f" ADF TEST RESULTS FACTOR {loc+1}: {series.name}"
  print("{:^150}".format(statement))
  print('\n')

  # Null Hypothesis Validation
  print("{:^150}".format('  APPRAOCH 1: Test Statistic value should be < Critical Test Statistic Value  '))
  ctsv = result[4]['1%']
  if result[0] < ctsv:
    statement = f"Test Statistic Value: {result[0]} < Critical Test Statistic Value: {ctsv}, {series.name} is Stationary."
    print("{:^150}".format(statement))
  else:
    statement = f"Test Statistic Value: {result[0]} > Critical Test Statistic Value: {ctsv}, {series.name} is Non-Stationary."
    print("{:^150}".format(statement))

  print('\n')

  # Null Hypothesis Validation
  print("{:^150}".format(" APPROACH 2: p-value should be <= 0.005 "))
  if result[1] <= 0.005:
    statement = f"P-value: {result[1]} is <= 0.005, {series.name} is Stationary."
    print("{:^150}".format(statement))
  else:
    statement = f"P-value: {result[1]} is > 0.005, {series.name} is Non-stationary."
    print("{:^150}".format(statement))
  print('\n\n')
```

```python
# Passing each factor to the adf_res function

for var in data_for_adf.columns:
  adf_res(data_for_adf[var], data_for_adf.columns.get_loc(data_for_adf[var].name))
```

Taking First difference and performing ADF for all variables

```python
# First Difference 

for var in data_for_adf.columns:
  adf_res(data_for_adf[var].diff().fillna(0), data_for_adf.columns.get_loc(data_for_adf[var].name))
```
